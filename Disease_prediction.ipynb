```python
data = pd.read_csv('newtest.csv')
che = pd.read_csv('chennai.csv')
mad = pd.read_csv('madurai.csv')
tan = pd.read_csv('tanjore.csv')
tot = pd.read_csv('total.csv')
decision = pd.read_csv('decision.csv')
```

```python
#leap year
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False
```
